```
gross_earn = dataset['gross_earn_clear']
```

```
max_value = gross_earn.max()
min_value = gross_earn.min()
print('Наибольший доход: ', max_value, 'Наименьший доход: ', min_value)
```

```
gross_earn = dataset[dataset['gross_earn_clear'] != 0.0]['gross_earn_clear']
```

```
max_value = gross_earn.max()
min_value = gross_earn.min()
print('Наибольший доход после очистки: ', max_value, 'Наименьший доход после очистки: ', min_value)
```

```
mean_value = gross_earn.mean()
median_value = gross_earn.median()
print('Средний доход с фильма: ', mean_value)
print('Медиана доходов фильмов: ', median_value)
```

```
percentile_10_value = gross_earn.quantile(0.1)
percentile_25_value = gross_earn.quantile(0.25)
percentile_50_value = gross_earn.quantile(0.5)
percentile_75_value = gross_earn.quantile(0.75)
percentile_90_value = gross_earn.quantile(0.9)

print(f'10-й процентиль: {percentile_10_value}, 25-й процентиль: {percentile_25_value}')
print(f'50-й процентиль: {percentile_50_value}, 75-й процентиль: {percentile_75_value}')
print(f'90-й процентиль: {percentile_90_value}')
```

```
gross_earn.describe()
```

```
import matplotlib.pyplot as plt
```

```
plt.boxplot(gross_earn.dropna())
plt.show()
```

```
plt.hist(gross_earn.dropna(), bins=20)
plt.show()
```

```
gross_earn_2 = dataset[
    (dataset['gross_earn_clear'] >= percentile_10_value) &
    (dataset['gross_earn_clear'] <= percentile_90_value)]['gross_earn_clear']
```

```
plt.hist(gross_earn_2, bins=20)
plt.show()
```

Вывод:

Размах доходов достаточно велик.
Присутствуют нулевые значения, которые были исключены из анализа.
Среднее значение доходов и медиана могут отличаться, что может свидетельствовать о наличии выбросов.
Процентили показывают, что 90% фильмов имеют доход меньше определенной величины, что также может помочь выявить выбросы.
Графики дают наглядное представление о распределении данных.

```
import seaborn as sns
```

```
sns.jointplot(x='runtime_clear', y='gross_earn_clear', data=dataset, kind='scatter', alpha=0.5)
plt.show()
```

```
columns = ['runtime_clear', 'gross_earn_clear']
g = sns.PairGrid(dataset[columns])
g.map(sns.scatterplot, alpha=0.5)
```

```
new_dataset = dataset.copy()
for col in columns:
    limiter = new_dataset[col].quantile(0.99)
    new_dataset = new_dataset[new_dataset[col] <= limiter]
```

```
g = sns.PairGrid(new_dataset[columns])
g.map(sns.scatterplot, alpha=0.5)
```

Вывод:

Визуально можно оценить наличие линейной или нелинейной взаимосвязи.
Выбросы могут влиять на результаты, поэтому их удаление помогает улучшить видимость связи между признаками.

```
directors = dataset['director'].unique()
len(directors)
```

```
directors_top_5 = dataset['director'].value_counts()[:5].index
```

```
for director in directors_top_5:
    df = dataset[dataset['director'] == director]
    certificate_counts = df['certificate'].value_counts()
    labels = certificate_counts.index
    colors = dict(zip(labels, plt.cm.tab20.colors[:len(labels)]))

    plt.title(f'{director}')
    plt.pie(certificate_counts, labels=labels, colors=[colors[label] for label in labels], autopct='%.1f%%')
    plt.show()
```

Вывод:

Анализ распределения категориальных признаков по разным режиссерам позволяет оценить, какой класс фильмов чаще всего создают топ-режиссеры.
